### 평가 지표: 정밀도, 재현율, F1 스코어
<div style="border:2px solid black; padding:10px; margin:10px;">
<div style="text-align: center;">
    <img src="../../../img/confusion_matrix.png" width="500"/><br>
</div>

머신 러닝 모델을 평가할 때, 특히 불균형한 데이터셋에서는 정확도만으로는 모델의 성능을 제대로 평가하기 어려운 경우가 있다.

#### 1. 정밀도 (Precision)
- 정밀도는 모델이 양성(Positive)으로 예측한 결과 중에서 실제로 양성인 비율을 의미한다.
- 즉, 모델이 사기 거래로 예측한 거래 중에서 실제로 사기인 거래의 비율
- 높은 정밀도를 갖는 모델은 거짓 양성(FP)을 줄이는 데 도움이 된다.
- 고객에게 불필요한 경고나 차단 등의 불편을 줄이기 위해 정밀도가 중요하다.

#### 2. 재현율 (Recall)
- 재현율은 실제 양성(Positive) 중에서 모델이 정확하게 양성으로 예측한 결과의 비율을 의미한다.
- 즉, 실제 사기 거래 중에서 모델이 얼마나 많은 사기 거래를 탐지하는지를 보여준다.
- 높은 재현율을 갖는 모델은 거짓 음성(FN)을 줄이는 데 도움이 된다.
- 사기 거래를 놓치지 않고 빠르게 대응하기 위해 재현율이 중요하다.

#### 3. F1 스코어 (F1 Score)
- F1 스코어는 정밀도와 재현율의 조화 평균으로 계산되며, 불균형한 데이터셋에서 모델의 성능을 평가하는데 유용하다.
- 높은 정밀도와 높은 재현율을 모두 갖는 모델일수록 F1 스코어가 높아진다.
- 정밀도와 재현율 사이의 균형을 제공하는 평가 지표로, 둘 중 어느 하나만 높고 다른 하나가 낮을 경우에도 상대적으로 높은 F1 스코어를 얻을 수 있다.

불균형한 데이터셋에서는 사기 탐지와 같이 중요한 문제에서 정밀도, 재현율, F1 스코어를 함께 고려하여 모델의 성능을 평가하는 것이 필요하다.<br> 이러한 평가 지표들을 이용하면 모델의 성능을 더 정확하게 파악하고, 사기 탐지와 같은 실제 문제에서 더 나은 결과를 얻을 수 있다.
</div>

###  최종 선택 평가 지표: ROC 곡선과 AUC (Area Under the Curve)

<div style="border:2px solid black; padding:10px; margin:10px;">
<div style="text-align: center;">
    <img src="../../../img/roc_curve.png" width="500"/><br>
</div>

머신 러닝 모델의 평가를 위해 ROC 곡선과 AUC (Area Under the Curve)는 불균형한 데이터셋에서 많이 사용되는 중요한 평가 지표이다.

### ROC 곡선 (Receiver Operating Characteristic Curve)
- ROC 곡선은 모델의 임계값을 변화시키면서, 모델의 Recall(TPR)에 대한 1-TPR(FPR)의 그래프이다.
- 임계값은 모델이 양성으로 판단하는 확률을 결정하는데, 이 값을 변화시키면서 얻는 재현율과 특이도의 변화를 나타낸다.
- ROC 곡선은 이진 분류 모델의 성능을 시각적으로 평가하는데 사용된다.
- ROC 곡선이 왼쪽 상단 모서리에 가까울수록 모델의 성능이 우수하다고 판단할 수 있다.

### AUC (Area Under the Curve)
- AUC는 ROC 곡선 아래의 면적을 나타낸다.
- AUC는 ROC 곡선의 형태와 면적으로 모델의 성능을 수치화한다.
- AUC 값은 0과 1 사이의 값을 가지며, 1에 가까울수록 모델의 성능이 우수하다.
- 무작위로 예측하는 모델의 AUC 값은 0.5이며, 이보다 큰 AUC 값은 모델이 유용하다고 판단한다.
- AUC 값이 0.5에 가까울 경우 모델의 성능이 좋지 않고, 0.5 미만이면 모델이 정밀도와 재현율 중 하나에 대해 나쁜 성능을 보이는 것을 의미한다.
</div>

```python
from sklearn.metrics import *
```

```python
y_pred_proba = lg_model.predict_proba(X_test)[:, 1]
```

```python
# ROC, AUC 값 구하기
roc_auc = roc_auc_score(y_test, y_pred_proba)

# ROC 커브 값 구하기
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

print('roc_auc: ',roc_auc,'\n')
print('-----------------------------')
print('fpr: ',fpr,'\n')
print('-----------------------------')
print('tpr: ',tpr,'\n')
```

```python
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# 반복 작업을 줄이기 위해 함수화
def X_y():
    df = pd.read_csv("assignment3_creditcard.csv")
    X = df.drop(['Class'], axis=1)
    y = df['Class']
    return X, y

def standard_scaling(X):
    scaler = StandardScaler()
    X_scaled  = scaler.fit_transform(X)
    return X_scaled

def logistic_model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
    lg_model = LogisticRegression()
    lg_model.fit(X_train, y_train)
    return lg_model, X_test, y_test
    
def ROC_AUC(lg_model, X_test, y_test):
    y_pred_proba = lg_model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    return roc_auc
```

### 1. 로그 변환

```python
X, y = X_y()
```

```python
# 데이터의 분포 확인 및 Scale 확인
X.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
X.describe()
```

```python
for col in X.columns:
    min_val = X[col].min()
    if min_val <= 0:
        X[col] = np.log1p(X[col] + np.abs(min_val) + 0.0001)
    else:
        X[col] = np.log1p(X[col])
```

```python
# 시각화 결과로는 로그 변환이 눈에 띄지 않지만, 수치로 보면 변했음이 확인 가능
X.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
X.describe()
```

```python
X_scaled = standard_scaling(X)
lg_model, X_test, y_test = logistic_model(X_scaled, y)
roc_auc = ROC_AUC(lg_model, X_test, y_test)
roc_auc
```

### 2. 데이터 불균형 처리(오버 샘플링)

### 4. 이상치 제거

```python
X, y = X_y()
```

```python
# 1주차에 사용한 코드 재활용
# 정규화 이후, ±3을 벗어나는 값 제거
from scipy.stats import zscore

variable_numeric_cols = df.columns[:-1]  # 'Class' column excluded

df_zscore = df.copy()

def handle_outliers_zscore(col, dataframe):
    z_scores = zscore(dataframe[col])
    abs_z_scores = np.abs(z_scores)

    outliers = dataframe[abs_z_scores > 3]
    outlier_count = outliers.shape[0]

    # Remove outliers
    dataframe.drop(outliers.index, inplace=True)

    return outlier_count

fig, axes = plt.subplots(len(variable_numeric_cols), 2, figsize=(14, len(variable_numeric_cols)*4))

for idx, col in enumerate(variable_numeric_cols):
    sns.distplot(df[col], ax=axes[idx, 0], color='blue', hist=False, kde=True)
    axes[idx, 0].set_title(f"Original {col} (Outliers: {handle_outliers_zscore(col, df_zscore)})", 
                            fontsize=14, pad=20)

    # Data after outlier handling
    sns.distplot(df_zscore[col], ax=axes[idx, 1], color='green', hist=False, kde=True)
    axes[idx, 1].set_title(f"After Z-Score Outlier Handling {col}", 
                            fontsize=14, pad=20)

plt.tight_layout()
plt.show()
```

```python
# 기존 df에서 Class가 1인 케이스 수를 카운트
class_1_count_original = df[df['Class'] == 1].shape[0]

# df_zscore 데이터프레임을 사용하여 이상치 처리 후 Class가 1인 케이스 수를 카운트
class_1_count_after_zscore = df_zscore[df_zscore['Class'] == 1].shape[0]

# 두 카운트 값의 차이를 계산하여 줄어든 Class가 1인 케이스 수를 확인
reduction_in_class_1_count = class_1_count_original - class_1_count_after_zscore

print(f"df(이상치 제거 전) 사기 데이터 수: {class_1_count_original}")
print(f"df_zscore(이상치 제거 후) 사기 데이터 수: {class_1_count_after_zscore}")
print(f"사기 데이터 감소량: {reduction_in_class_1_count}")
```

```python
X_scaled = df_zscore.drop(['Class'], axis=1)
y = df_zscore['Class']
lg_model, X_test, y_test = logistic_model(X_scaled, y)
roc_auc = ROC_AUC(lg_model, X_test, y_test)
roc_auc
```
