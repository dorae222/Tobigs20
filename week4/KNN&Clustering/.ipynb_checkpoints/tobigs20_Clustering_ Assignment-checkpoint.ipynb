### 범주형 변수 더미화

```python
# 범주형 변수 더미화
dummy_data = pd.get_dummies(df, columns=['Gender'])

# 더미화 후 데이터의 첫 5행 출력
dummy_data.head()
```

### 모든 컬럼을 스케일링

```python
from sklearn.preprocessing import StandardScaler

# 표준 스케일러 초기화
scaler = StandardScaler()

# 데이터프레임에 스케일링 적용
dummy_data = pd.DataFrame(scaler.fit_transform(dummy_data), columns=dummy_data.columns)
```

```python
columns_to_visualize = dummy_data.columns

plt.figure(figsize=(15,10))
for i, col in enumerate(columns_to_visualize, 1):
    plt.subplot(2, 3, i)
    sns.histplot(dummy_data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

```python
sns.set_style("whitegrid")

plot = sns.pairplot(dummy_data)
plt.show()
```

## 2. Clustering

### KMEANS

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# KMeans 군집화를 위한 최적의 군집 수 찾기
range_n_clusters = list(range(2,11))
silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(dummy_data)
    
    # 실루엣 점수 계산
    silhouette_avg = silhouette_score(dummy_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# 최적의 군집 수를 실루엣 점수로부터 결정
optimal_clusters = range_n_clusters[silhouette_scores.index(max(silhouette_scores))]

optimal_clusters, max(silhouette_scores)
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN

eps_values = [0.3, 0.5, 0.7, 0.9]
min_samples_values = [3, 5, 7]
dbscan_results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        cluster_labels = dbscan.fit_predict(dummy_data)
        
        # 실루엣 점수 계산 (군집이 1개 또는 전체 데이터 포인트가 노이즈일 경우 제외)
        if len(set(cluster_labels)) > 1 and -1 in cluster_labels:
            silhouette_avg = silhouette_score(dummy_data, cluster_labels)
            dbscan_results.append((eps, min_samples, silhouette_avg))

# 최적의 eps와 min_samples 값을 실루엣 점수로부터 결정
optimal_eps, optimal_min_samples, optimal_silhouette = max(dbscan_results, key=lambda x: x[2])

optimal_eps, optimal_min_samples, optimal_silhouette
```

## 3. Evaluation

### DUNN Index

```python
from scipy.spatial import distance

def dunn_index(X, cluster_labels):
    # 군집 내의 최대 거리 계산
    clusters = set(cluster_labels)
    intra_cluster_max_distance = []
    for cluster in clusters:
        cluster_points = X[cluster_labels == cluster]
        if len(cluster_points) > 1:  # 하나 이상의 포인트가 있는 군집에 대해서만
            intra_distance = distance.pdist(cluster_points).max()
            intra_cluster_max_distance.append(intra_distance)
    
    max_intra_cluster_distance = max(intra_cluster_max_distance)
    
    # 군집 간의 최소 거리 계산
    inter_cluster_min_distance = []
    for cluster_i in clusters:
        for cluster_j in clusters:
            if cluster_i != cluster_j:
                distance_between_clusters = distance.cdist(X[cluster_labels == cluster_i], 
                                                          X[cluster_labels == cluster_j]).min()
                inter_cluster_min_distance.append(distance_between_clusters)
    
    min_inter_cluster_distance = min(inter_cluster_min_distance)
    
    # Dunn 지수 계산
    return min_inter_cluster_distance / max_intra_cluster_distance
```

```python
# KMeans와 DBSCAN에 대한 Dunn 지수 계산
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(dummy_data)

dbscan = DBSCAN(eps=optimal_eps, min_samples=optimal_min_samples)
dbscan_labels = dbscan.fit_predict(dummy_data)

kmeans_dunn = dunn_index(dummy_data.values, kmeans_labels)
dbscan_dunn = dunn_index(dummy_data.values, dbscan_labels)

kmeans_dunn, dbscan_dunn
```

```python
from sklearn.decomposition import PCA

# 3차원으로 PCA 차원 축소
pca_3d = PCA(n_components=3)
pca_result_3d = pca_3d.fit_transform(dummy_data)
```

```python
# Seaborn 스타일 설정
sns.set_style("whitegrid")

fig = plt.figure(figsize=(15, 7))

# KMeans 결과 3D 시각화
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
scatter1 = ax1.scatter(pca_result_3d[:, 0], pca_result_3d[:, 1], pca_result_3d[:, 2], 
                       c=kmeans_labels, cmap='tab10', s=40)
legend1 = ax1.legend(*scatter1.legend_elements(num=10), loc="upper right", title="Clusters")
ax1.add_artist(legend1)
ax1.set_title('KMeans Clustering (n_clusters=10)')

# DBSCAN 결과 3D 시각화
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
scatter2 = ax2.scatter(pca_result_3d[:, 0], pca_result_3d[:, 1], pca_result_3d[:, 2], 
                       c=dbscan_labels, cmap='tab10', s=40)
legend2 = ax2.legend(*scatter2.legend_elements(), loc="upper right", title="Clusters")
ax2.add_artist(legend2)
ax2.set_title('DBSCAN Clustering (eps=0.9, min_samples=5)')

plt.tight_layout()
plt.show()
```
