# 1. EDA

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
path = './ml-latest-small/'
movies_df = pd.read_csv(os.path.join(path + 'movies.csv'), index_col='movieId', encoding='utf-8')
```

```python
movies_df.head()
```

## 1. [movies.csv] 장르 분포 확인

```python
genre_lists = [x.split('|') for x in movies_df['genres'].values]

unique_genres = set()
genre_counts = {}

for genre_list in genre_lists:
    unique_genres.update(genre_list)
    for genre in genre_list:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1
    
unique_genres_list = list(unique_genres)
unique_genres_list
```

```python
# 장르 분포 시각화
plt.figure(figsize=(15, 7))
plt.bar(genre_counts.keys(), genre_counts.values(), color='royalblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('genre')
plt.ylabel('genre_counts')
plt.title('Distribution of the number of movies by genre')
plt.tight_layout()
plt.show()
```

## 2. [movie.csv] Title 기준 연도별 분포 확인

```python
# 연도 추출
movies_df['year'] = movies_df['title'].str.extract(r'\((\d{4})\)')

# 연도 기준 카운팅
year_counts = movies_df['year'].value_counts().sort_index()
```

```python
# 연도별 분포 시각화
plt.figure(figsize=(10, 6))
movies_df['year'].dropna().astype(int).hist(bins=50, color='royalblue', rwidth=0.9)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Counts of movies by year')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 3. [movie.csv] 연도별, 장르 분포 확인하기

```python
for genre in unique_genres_list:
    movies_df[genre] = movies_df['genres'].str.contains(genre).astype(int)

genre_by_year = movies_df.groupby('year').sum()

plt.figure(figsize=(10, 6))
genre_by_year[unique_genres_list].plot(kind='bar', stacked=True, figsize=(20, 10))
plt.xlabel('Year')
plt.ylabel('Genre Count')
plt.title('Genre Count by Year')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

# 2. Memory-based CF

```python
links_df = pd.read_csv(os.path.join(path + 'links.csv'), encoding='utf-8')
movies_df = pd.read_csv(os.path.join(path + 'movies.csv'), encoding='utf-8')
ratings_df = pd.read_csv(os.path.join(path + 'ratings.csv'), encoding='utf-8')
tags_df = pd.read_csv(os.path.join(path + 'tags.csv'), encoding='utf-8')
```

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split

# Reader를 설정합니다. 평점의 범위는 0.5부터 5까지입니다.
reader = Reader(rating_scale=(0.5, 5.0))

# ratings.csv 데이터를 Surprise의 Dataset 형식으로 로드합니다.
data = Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], reader)

# 데이터를 학습 데이터와 테스트 데이터로 분할합니다. (테스트 데이터의 비율은 20%로 설정합니다.)
trainset, testset = train_test_split(data, test_size=0.2)
```

## KNN-BASIC

```python
from surprise import KNNBasic
from surprise.accuracy import rmse

knn_basic = KNNBasic()
knn_basic.fit(trainset)
predictions_knn_basic = knn_basic.test(testset)
rmse_knn_basic = rmse(predictions_knn_basic)
```

## KNN-with-ZScore

```python
from surprise import KNNWithZScore

knn_zscore = KNNWithZScore()
knn_zscore.fit(trainset)
predictions_knn_zscore = knn_zscore.test(testset)
rmse_knn_zscore = rmse(predictions_knn_zscore)
```

# 3. Model-based CF(SVD++)

```python
from surprise import SVDpp
from surprise.accuracy import rmse
from surprise.model_selection import train_test_split

# SVD++ 모델 설정 및 학습
svdpp = SVDpp()
svdpp.fit(trainset)
predictions_svdpp = svdpp.test(testset)

# RMSE 계산
rmse_svdpp = rmse(predictions_svdpp)
```

```python
from surprise.model_selection import GridSearchCV

# 튜닝할 하이퍼파라미터 설정
param_grid = {
    'n_epochs': [5, 10, 20],
    'lr_all': [0.002, 0.005, 0.01],
    'reg_all': [0.2, 0.4, 0.6]
}

# GridSearchCV를 사용하여 최적의 하이퍼파라미터를 찾습니다.
gs = GridSearchCV(SVDpp, param_grid, measures=['rmse'], cv=3)
gs.fit(data)

# 최적의 하이퍼파라미터와 해당 RMSE 값 출력
best_params = gs.best_params['rmse']
best_score = gs.best_score['rmse']

best_params, best_score
```

- surprise에서는 tqdm이 지원되지 않아서, 시간 파악이 안돼 grid search 걸어놓고 기다리느라 깊게 분석하지 못해서 아쉽습니다...

- KNN-BASIC의 성능이 제일 좋긴한데, 추천에 대한 이해도 깊이 부족으로 정확하게 왜 그러한지는 파악하지 못했습니다...
    - 당연히 Model 기반이 우수할 줄 알았으나, 예상과 다르게 기본 알고리즘이 가장 성능이 좋게 나왔습니다.
